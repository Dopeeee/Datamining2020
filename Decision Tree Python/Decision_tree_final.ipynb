```python
def find_max(Set,index):
    max_num = Set[0][index]
    for i in range(len(Set)):
        if max_num < Set[i][index]:
            max_num = Set[i][index]
    return max_num
```

```python
# 输出对于一个ordinal数据，有多少种划分方法。（从小到大排序）
def separate_method_ordinal(Set_before,index):
    Set_A = []
    for i in range(len(Set_before)):
        if(Set_before[i][index] in Set_A):
            pass
        else:
            Set_A.append(Set_before[i][index])
    Set_A.sort()
    return Set_A
```

```python
# 查找这个集合里有多少个1，对应公式里的P_y， P_n的数量就是len(set_a) - num_of1
def find_1_num(set_a):
    num_of1 = 0
    for i in range(len(set_a)):
        if set_a[i][14] == 1:
            num_of1 = num_of1 + 1
    return num_of1
```

```python
# GINI range from 0 to 0.5 , smaller means better
def cal_gini_set(set_a):
    label_1_num = find_1_num(set_a)
    label_0_num = len(set_a) - label_1_num 
    total_num = label_1_num + label_0_num
    label_1_percentage = label_1_num / total_num
    label_0_percentage = 1 - label_1_percentage
    # GINI = 1 - (P_y平方 + P_n平方)
    GINI = 1 - (label_1_percentage*label_1_percentage + label_0_percentage* label_0_percentage )
    return GINI
```

```python
gini_original_set = cal_gini_set(points_list)
gini_original_set
```

```python
# GINI of the split
# Same with the Gini of a set, The GINI of the split is also smaller means better
def cal_gini_split(set_a,set_b):
    Gini_a = cal_gini_set(set_a)
    #print(Gini_a)
    Gini_b = cal_gini_set(set_b)
    #print(Gini_b)
    a_num = len(set_a)
    b_num = len(set_b)
    total_num = a_num + b_num
    Gini_split = (a_num/total_num) * Gini_a + (b_num/total_num) * Gini_b
    return Gini_split
```

```python
# 我们写了separate_method_ordinal， 举出了所有ordinal的分类方法
# 现在需要写一个函数，根据ordinal的分类方法 ，构建两个分类的数据集合
def construct_set(set_total,set_construct,index):
    # 构建数据集，用来存储新的分类结果
    # a集合满足这个set_construct
    # b集合不满足，也就是互斥的集合。（反正是二叉决策树）
    new_set_a = []
    new_set_b = []
    for i in range(len(set_total)):
        #如果这个元素，存在set_construct里，就Append进new_set_a
        if set_total[i][index] in set_construct:
            new_set_a.append(set_total[i])
        else:
            new_set_b.append(set_total[i])
    return new_set_a,new_set_b
```

```python
# 构建关于不可排序的feature的所有可能性集合。 传入的diff_set是包含的所有不重复的元素，可以理解为domain。
def construct_nornimal_possible_set(diff_set):
    possible_set = []
    lenth_num = len(diff_set)
    lenth_num = math.floor(lenth_num/2)
    for i in range(lenth_num):
        possible_set = possible_set + list(combinations(diff_set, i+1))
    return possible_set
```

```python
# 针对某一个值（index,目前是ordinal. 如果是nominal，后面可以写进去)，穷举法算GINI
def find_best_GINI(set_a, index):
    tmp_set = separate_method_ordinal(set_a,index)
    smallest_gini = 1
    best_a = []
    best_b = []
    best_precise = []
    # 对于可排序的int类型：
    if( type(set_a[0][index]) == int ):
        for i in range(len(tmp_set)-1):
            # 这里-1 是为了避免最后有空集
            tmp_a = tmp_set[:i+1]
            # 这里的tmp_b 应该是可以删除的
            tmp_b = tmp_set[i+1:]
            #print(tmp_a)
            #print(tmp_b)
            set_tmp_a = []
            set_tmp_b = []
            set_tmp_a,set_tmp_b = construct_set(set_a,tmp_a,index)
            this_gini = cal_gini_split(set_tmp_a,set_tmp_b)
            if this_gini < smallest_gini:
                smallest_gini = this_gini
                best_a = set_tmp_a.copy()
                best_b = set_tmp_b.copy()
                best_precise = tmp_a
    # 对于不可排序的str类型
    elif( type(set_a[0][index]) == str ):
        possible_set = construct_nornimal_possible_set(tmp_set)
        #遍历整个possible_set，开始找最小的gini
        for i in range(len(possible_set)):
            tmp_a = possible_set[i]
            set_tmp_a = []
            set_tmp_b = []
            set_tmp_a,set_tmp_b = construct_set(set_a,tmp_a,index)
            this_gini = cal_gini_split(set_tmp_a,set_tmp_b)
            if this_gini < smallest_gini:
                smallest_gini = this_gini
                best_a = set_tmp_a.copy()
                best_b = set_tmp_b.copy()
                #返回断言，也就是分割标准
                best_precise = tmp_a
                
    return smallest_gini,best_a,best_b,best_precise
```
